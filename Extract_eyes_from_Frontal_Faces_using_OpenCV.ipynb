### Load cascade Classifier and view images

```
path_data = '/content/drive/My Drive/iris detection/data/'
path_img = '/content/drive/My Drive/iris detection/images/'

face_cascade = cv2.CascadeClassifier(path_data + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(path_data + 'haarcascade_eye.xml')
```

```
img = cv2.imread(path_img + "mew_face.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```
cv2_imshow(cv2.resize(gray,(384,240)))
```

### Crop detected eyes and save them

```
faces = face_cascade.detectMultiScale(gray, 1.2, 5) # image, scale factor, neighbours
cropped_eyes=[]

for (x,y,w,h) in faces:
  img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
  roi_gray = gray[y:y+h, x:x+w]
  roi_color = img[y:y+h, x:x+w]
  eyes = eye_cascade.detectMultiScale(roi_gray,1.6,5)
  for (ex,ey,ew,eh) in eyes:
    cropped_eyes.append(roi_color[ey:ey+eh, ex:ex+ew])
    cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```

```
cv2_imshow(cv2.resize(img,(512,320)))
```

```
cropped_eyes = np.array(cropped_eyes)
```

```
cv2.imwrite((path_img + 'mew_iy1.jpg'), cropped_eyes[0])
```

```
cv2.imwrite((path_img + 'mew_iy2.jpg'), cropped_eyes[1])
```
